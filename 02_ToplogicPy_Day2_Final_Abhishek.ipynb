```python
from topologicpy.Vertex import Vertex
from topologicpy.Edge import Edge
from topologicpy.Wire import Wire
from topologicpy.Face import Face
from topologicpy.Shell import Shell
from topologicpy.Cell import Cell
from topologicpy.CellComplex import CellComplex
from topologicpy.Cluster import Cluster
from topologicpy.Topology import Topology
from topologicpy.Dictionary import Dictionary
from topologicpy.Graph import Graph
```

```python
building = Topology.ByOBJPath(r"D:\AS-build\eCAADeTopologicPyAS\myworkingfiles\Abhishek\Abhishek\Building.obj", transposeAxes=True)[0]
faces = Topology.Faces(building)
building = Cell.ByFaces(faces, tolerance=0.001)
building = Topology.RemoveCoplanarFaces(building)
Topology.Show(building,renderer="browser")
```

```python
f = Face.Rectangle(origin=Topology.Centroid(building), width=50, length=50)
f1 = Topology.Translate(f, 0, 0, -1)
f2 = Topology.Translate(f1, 0, 0, 3)
f3 = Topology.Rotate(f, origin=Topology.Centroid(f), axis=[0,1,0], angle=90)
f4 = Topology.Translate(f3, -5, 0, 0)
f5 = Topology.Translate(f4, -4, 0, 0)
f6 = Topology.Translate(f3, 5, 0, 0)
f7 = Topology.Translate(f6, 4, 0, 0)
f8 = Topology.Translate(f7, 5, 0, 0)
f9 = Topology.Rotate(f, origin=Topology.Centroid(f), axis=[1,0,0], angle=90)
f10 = Topology.Translate(f9, 0, 4, 0)
f11 = Topology.Translate(f10, 0, 2.5, 0)
f12 = Topology.Translate(f9, 0, -2, 0)
cluster = Cluster.ByTopologies(f1, f2, f3, f4, f5, f6, f7, f8, f9, f10, f11, f12)
building = Topology.Slice(building, cluster)
Topology.Show(building,renderer="browser")
```

```python
core = Topology.ByOBJPath(r"D:\AS-build\eCAADeTopologicPyAS\myworkingfiles\Abhishek\Abhishek\Core.obj", transposeAxes=True)[0]

core = Topology.SelfMerge(core)
core = Topology.RemoveCoplanarFaces(core)
core = Topology.SelfMerge(core)

# Topology.Show(core)
# core = Cell.ByFaces(faces, tolerance=0.001)
# core = Topology.RemoveCoplanarFaces(core)
# Topology.Show(core)
```

```python
building = Topology.Difference(building, core)
#Topology.Show(building)
g = Graph.ByTopology(building)
#Graph.Show(g)
```

```python
columns = Topology.ByOBJPath(r"D:\AS-build\eCAADeTopologicPyAS\myworkingfiles\Abhishek\Abhishek\Columns.obj", transposeAxes=True)[0]
print(columns)
columns = Topology.SelfMerge(columns)
columns = Topology.RemoveCoplanarFaces(columns)
columns = Topology.SelfMerge(columns)
print(Topology.Analyze(columns))
#Topology.Show(building, core, columns)
```

```python
ground = Topology.ByOBJPath(r"D:\AS-build\eCAADeTopologicPyAS\myworkingfiles\Abhishek\Abhishek\Ground.obj", transposeAxes=True)[0]
print(ground)
ground = Topology.SelfMerge(ground)
print(ground)
ground = Topology.Slice(ground, cluster)
#Topology.Show(ground)
```

```python
ground = Topology.RemoveCoplanarFaces(ground)
print(ground)
#Topology.Show(ground)
```

```python
Topology.Show(building, core, columns, ground, renderer="browser")
```

```python
selectors = []
bldg_cells = Topology.Cells(building)
core_cells = Topology.Cells(core)
column_cells = Topology.Cells(columns)
ground_cells = Topology.Cells(ground)
```

```python
d = Dictionary.ByKeyValue("label", 3)
for bldg_cell in bldg_cells:
    v = Topology.InternalVertex(bldg_cell)
    v = Topology.SetDictionary(v, d)
    selectors.append(v)

d = Dictionary.ByKeyValue("label", 4)
for core_cell in core_cells:
    v = Topology.InternalVertex(core_cell)
    v = Topology.SetDictionary(v, d)
    selectors.append(v)

d = Dictionary.ByKeyValue("label", 2)
for column_cell in column_cells:
    v = Topology.InternalVertex(column_cell)
    v = Topology.SetDictionary(v, d)
    selectors.append(v)

d = Dictionary.ByKeyValue("label", 0)
for ground_cell in ground_cells:
    v = Topology.InternalVertex(ground_cell)
    v = Topology.SetDictionary(v, d)
    selectors.append(v)
```

```python
all_cells = bldg_cells + core_cells + column_cells + ground_cells
result = CellComplex.ByCells(all_cells)
print(result)
Topology.Show(result,renderer="browser")
```

```python
result = Topology.TransferDictionariesBySelectors(result, selectors, tranCells=True)
```

```python
g = Graph.ByTopology(result)
Graph.Show(g, vertexGroupKey="label", vertexGroups=[0,2,3,4], vertexLabelKey="label",renderer="browser")
```
